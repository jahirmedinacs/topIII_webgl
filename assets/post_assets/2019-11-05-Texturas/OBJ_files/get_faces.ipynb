```python
import numpy as np
import re

def load_obj(path):
    ref = open(path, "r")

    vertices = []
    indices = []

    for line in ref.readlines():
        if line[0] == "v" or "f":
            temp = re.sub(' +', ' ', line[1:])
            temp = temp.rstrip()

            if line[0] == "v":
                temp = [float(ii) for ii in temp.split(" ")[1:]]
                vertices += temp
            if line[0] == "f":
                temp = [int(ii.split("/")[0]) for ii in temp.split(" ")[1:]]
                indices += [temp]
        else:
            pass

    return [vertices, indices]


[obj_vtx, obj_idx] = load_obj("hexagon-faces.obj")
```

```python
ii = 1
for face in obj_idx:
    print("// FACE {:d}".format(ii))
    for vtx in face:
        print(np.round(obj_vtx[vtx - 1], 5), end=", ")
    print("")
    ii += 1
```
