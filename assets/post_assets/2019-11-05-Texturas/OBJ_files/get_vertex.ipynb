```python
import numpy as np
import re

def load_obj(path):
    ref = open(path, "r")

    vertices = []
    indices = []

    for line in ref.readlines():
        if line[0] == "v" or "f":
            temp = re.sub(' +', ' ', line[1:])
            temp = temp.rstrip()

            if line[0] == "v":
                temp = [float(ii) for ii in temp.split(" ")[1:]]
                vertices += temp
            if line[0] == "f":
                temp = [int(ii.split("/")[0]) - 1 for ii in temp.split(" ")[1:]]
                indices += temp
        else:
            pass

    return [vertices, indices]


[obj_vtx, obj_idx] = load_obj("hexagon-volumen.obj")
```

```python
ll = 1
for ii in obj_vtx:
    if ll % 3 == 0:
        print(" {:f},".format(ii))
        if ll % 12 == 0:
            print("// face {:.0f}".format(ll / 12 + 1))
    else:
        print(" {:f},".format(ii), end="")
    ll += 1
```

```python
ll = 1
for ii in obj_idx:
    if ll % 12 == 0:
        print(" {:d},".format(ii))
        print("// face {:.0f}".format(ll / 18 + 1))
    else:
        print(" {:d},".format(ii), end="")
    ll += 1
```
